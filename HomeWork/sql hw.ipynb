```python
df = spark.read.format("csv").option("header","true").load("Files/data/Supermart Grocery Sales - Retail Analytics Dataset.csv")
display(df)
```

```python
df.createOrReplaceTempView("orders")
```

```python
%%sql
SELECT * FROM orders
```

1.**Write a query to find the total sales in each region**

```python
%%sql
SELECT Region, COUNT(*) AS Total_sales
FROM orders 
GROUP BY Region

```

**2.Count how many orders were placed by each customer.**

```python
%%sql
SELECT  Customer_Name , COUNT(Order_ID) AS Total_Orders 
FROM orders 
GROUP BY Customer_Name
```

**4.List the top 3 orders by profit.**

```python
%%sql
SELECT profit FROM orders ORDER BY Profit DESC limit 3
```

**3.Find all orders with a discount of more than 20% but less than 30%.**

```python
%%sql
SELECT Discount FROM orders WHERE (Discount>0.2) AND (Discount<0.3)
```

**5.Replace the category 'Snacks' with 'Quick Bites' in the result se**

```python
%%sql
SELECT Order_ID, 
       REPLACE(Category, 'Snacks', 'Quick Bites') AS category 
FROM orders;

```
